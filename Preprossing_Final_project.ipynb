```python
import pandas as pd

df = pd.read_csv("superstore_clean.csv", encoding='ISO-8859-1')


print("Data preview:")
print(df.head())


print("\nColumns in dataset:")
print(df.columns)
```

```python
# Display general info: data types, non-null counts
print("\nData Info:")
print(df.info())


print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())

# Option 1: Remove rows with missing values (if they are few)
#df = df.dropna()

# Option 2: Fill missing values (for example, forward fill or fill with a default value)
# df.fillna(method='ffill', inplace=True)
```

```python
# Remove duplicate rows based on a subset of columns, for example, Order ID, Customer ID, and Product ID
df = df.drop_duplicates(subset=["Order ID", "Customer ID", "Product ID"])

# Verify duplicates removed
print("\nDuplicates after cleaning:")
print(df.duplicated().sum())
```

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Create a new column 'Delivery Days' which calculates the difference between Ship Date and Order Date
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Verify changes
print("\nData after date conversion and feature creation:")
print(df[['Order Date', 'Ship Date', 'Delivery Days']].head())
```
